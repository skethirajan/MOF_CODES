```python
import os
from ase.io import read, Trajectory
import glob
import re

os.chdir("/global/u2/s/sudheesh/mofs_1/dft")

def is_converged(dir):
    start_dir = os.getcwd()
    os.chdir(dir)
    for line in open(f"OUTCAR", "r"):
        if re.search("accuracy", line):
            os.chdir(start_dir)
            return True
    os.chdir(start_dir)
    return False

def append_structures(dir):
    start_dir = os.getcwd()
    os.chdir(dir)
    vasprun_files = [file for file in os.listdir() if "vasprun_" in file]
    vasprun_files.sort(key=lambda x: int(re.split("[_.]", x)[1]))
    vasprun_files.append("vasprun.xml")
    atoms_list = []
    traj = Trajectory("vasp_relax.traj", "a")

    for file in vasprun_files:
        atoms_list.extend(read(file, index=":", format="vasp-xml"))

    if is_converged("./"):
        for atoms in atoms_list:
            traj.write(atoms)
    else:
        for atoms in atoms_list[:-1]: ## not writing the last index from vasprun.xml file
            traj.write(atoms)
    
    os.chdir(start_dir)

def delete_files(dir, ignore_files:list, keep_vasp_files=True):
    start_dir = os.getcwd()
    os.chdir(dir)

    if keep_vasp_files:
        vasp_files = ["CHG", "CHGCAR", "CONTCAR", "DOSCAR", "EIGENVAL", "IBZKPT", "INCAR", "KPOINTS", \
                      "OSZICAR", "OUTCAR", "PCDAT", "POSCAR", "POTCAR", "PROCAR", "REPORT", "WAVECAR", "XDATCAR", \
                      "ase-sort.dat", "vasp.out", "vasprun.xml"]
        ignore_files.extend(vasp_files)
        
    files = [file for file in os.listdir() if file not in ignore_files and ".sh" not in file and ".py" not in file]

    for file in files:
        os.remove(file)
    os.chdir(start_dir)

def get_nsw(dir):
    start_dir = os.getcwd()
    os.chdir(dir)
    try:
        atoms = read("vasp_relax.traj", index=":")
        nsw = len(atoms)
    except:
        nsw = "N/A"
    os.chdir(start_dir)
    return nsw

"""
search_dirs = []
for dir in glob.glob("./mof-*_linker-*/*/*/"):
    if "aimd" in dir:
        continue
    elif "ts0_relax" in dir:
        continue
    elif "linker-7" in dir:
        continue
    else:
        search_dirs.append(dir)
search_dirs.sort()
"""

# for index, dir in enumerate(search_dirs):
#     print(f"{index+1}): \t {dir} \t {is_converged(dir)} \t nsw = {get_nsw(dir)}")

"""
for dir in search_dirs:
    if not is_converged(dir):
        print(f"{dir} \t nsw = {get_nsw(dir)}")
"""

"""
search_dirs = [
    "./mof-4_linker-4(4)/relaxation/isif=2/", 
    "./mof-5_linker-5(1)/diffusion_neb/fs_relax/",
    "./mof-5_linker-5(4)/relaxation/isif=2/", 
    "./mof-6_linker-6(1)/diffusion_neb/fs_relax/",
    "./mof-6_linker-6(2)/relaxation/isif=2/", 
    "./mof-6_linker-6(4)/relaxation/isif=2/"]
"""

"""
for index, dir in enumerate(search_dirs):
    print(f"{index+1}): \t {dir}")
    append_structures(dir)
    delete_files(dir, ignore_files=["vasp_relax.traj", "job.err", "job.out"], keep_vasp_files=True)
"""

"""
for dir in search_dirs:
    if not is_converged(dir):
        print(f"{dir} \t nsw = {get_nsw(dir)}")
"""

# dir = "./mof-6_linker-6(1)/diffusion_neb/ts_relax/"
# append_structures(dir)
# delete_files(dir, ignore_files=["vasp_relax.traj", "ts0.traj", "OUTCAR"], keep_vasp_files=False)

search_dirs = ["./mof-2_linker-2(1)/diffusion_neb/ts_relax",
"./mof-3_linker-3(1)/diffusion_neb/ts_relax",
"./mof-4_linker-4(1)/diffusion_neb/ts_relax",
"./mof-5_linker-5(1)/diffusion_neb/ts_relax",
"./mof-6_linker-6(1)/diffusion_neb/ts_relax"]

for dir in search_dirs:
    print(f"{dir}: {get_nsw(dir)}")
```
